## Verificando informação inicial dos dados:

```python
def dataset_info(dataset_name: str, df: pd.DataFrame):
    print(f'Informações sobre o dataset {dataset_name}:', end='\n\n')
    print(f'Dataset Shape: {df.shape[0]} rows and {df.shape[1]} columns', end='\n\n')
    print(f'Total de missing values: {df.isnull().sum().sum()}', end='\n\n')
    print(f'Total de colunas com missing values: {df.isnull().any().sum()}', end='\n\n')
    print(f'Missing values por coluna:\n{df.isnull().sum()}', end='\n\n')
    print(f'Total de linhas duplicadas: {df.duplicated().sum()}', end='\n\n')
    print(df.head())
    
```

```python
datasets = {'RegistoTempos': df_tempos,
           'Tarefas': df_tarefas,
           'Clientes': df_clientes,
           'Colaboradores': df_colaboradores}
```

```python
for name, dataset in datasets.items():
    dataset_info(name, dataset)
    print('\n' + '-'*90 + '\n')
```
